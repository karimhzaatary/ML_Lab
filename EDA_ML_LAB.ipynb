```python
data = pd.read_csv('Orders.csv',encoding = "ISO-8859-1",index_col = False)
data1 = data[:10]
data.Profit = data.Profit.map(lambda x: int(x.replace("$","").replace(".","").replace(",",""))/100)
data.Sales = data.Sales.map(lambda x: int(x.replace("$","").replace(".","").replace(",",""))/100)
data
```

```python
data1['Order.Date']
data['Order.Date'] = data['Order.Date'].map(lambda x: datetime.strptime(x,'%m/%d/%y'))
#data1['Order.Date'] = date(data1['Order.Date'])
```

```python
data['Month'] = 2
data['Year'] = 1
data['Month'] = data['Order.Date'].map(lambda x: x.strftime("%B"))
data['Year'] = data['Order.Date'].map(lambda x: x.strftime("%Y"))
```

```python
month_ = data.groupby('Month')
```

```python
x = month_['Quantity'].agg(sum)
x.plot.bar()
```
